## Matplotlib

`A plotting library`

`Allows us to turn our data to plots or figures`

`1. Built on NumPy arrays (and Python)
 2. Integrates directly with pandas
 3. Can create basic or advanced plots
 4. Simple to use interface`

`1. Create Data
 2. Create plot(figure)
 3. Plot data(axes on figure)
 4. Customise plot`

```python
# %matplotlib inline # Magic Command that tells jupyter notebook to plot and visualisations within the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Recommended method (Object Oriented API)

```python
# Third way
fig, ax = plt.subplots() # subplots: more than 1 plot coming out at the same time
ax.plot(x, y);
type(fig), type(ax)
# (matplotlib.figure.Figure - The big blue square below, 
# matplotlib.axes._subplots.AxesSubplot: the plot, they are always the subplot)
```

### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # figsize: (width and height), and it is usually x then y - dimension of plot

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')

# 5. Save & show (Save the whole figure)
fig.savefig('saved_images/sample-plot.png')
```

```python
# Create some data
x = np.linspace(0, 10, 100) # Return evenly spaced number between a interval (Start, Stop, Number)
x[:10]
```

```python
# Make a plot from dictionary, Bar Plot
nut_butter_prices = {'Almond': 10, 'Peanut Butter': 8, 'Cashew Butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values()) # (x, height)
ax.set(title='Nut Butter Store', xlabel='Butter Type', ylabel='Price ($)');
```

```python
# Horizontal Bar Graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title='Nut Butter Store', xlabel='Prices ($)', ylabel='Butter Type');
```

```python
# Histogram
x = np.random.randn(1000) # Normal Distribution curve is like a symmetrical curve, and data fit within
fig, ax = plt.subplots()
ax.hist(x);
```

### Multiple plots at the same time - Subplots

```python
# Option 1

# 2 Rows, 2 Columns: Total of 4 graph spaces
# ax1, ax2, ax3, ax4: Total of 4 graphs
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axes (index)
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axes (index)
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Plotting from Pandas DataFrame (Object Oriented Method)

* When plotting something quickly, it is okay to use the pyplot method
* When plotting something more advanced, use the Object Oriented method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
# Object-Oriented Method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
# ax.set_xlim([45, 100]); # Set the range of x-axis
```

```python
# Object-Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],  y=over_50['chol'], c=over_50['target'])

# Customize
ax.set(title='Heart Disease & Cholesterol Levels', xlabel='Age', ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target'); # Grab the values from scatter, unpack lengend

# Add a horizontal line across the mean cholesterol
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True) # Creating a single figure, with 2 axes
# Sharex -> Share the same x axis value

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# Customize ax0
ax0.set(title='Heart Disease & Cholesterol Levels', ylabel='Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

# Customize ax1
ax1.set(title='Heart Disease & Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Add a legennd to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');

# Add the title to the figure (fig)
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
# Change the style within an existing style

# Set the style
plt.style.use('seaborn-whitegrid')

# Object-Oriented method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],  y=over_50['chol'], c=over_50['target'], cmap='winter') # cmap changes the colour scheme

# Customize
ax.set(title='Heart Disease & Cholesterol Levels', xlabel='Age', ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target'); # Grab the values from scatter, unpack lengend

# Add a horizontal line across the mean cholesterol
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
# Customizing the y and x axis limitations

# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True) # Creating a single figure, with 2 axes
# Sharex -> Share the same x axis value

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')

# Customize ax0
ax0.set(title='Heart Disease & Cholesterol Levels', ylabel='Cholesterol')

# Change the x axis limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(y=over_50['chol'].mean(), linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')

# Customize ax1
ax1.set(title='Heart Disease & Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')

# Change ax1 x-axis limits
ax1.set_xlim([50,80])

# Change ax1 y-axis limits
ax1.set_ylim([60,200])

# Add a legennd to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');

# Add the title to the figure (fig)
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
